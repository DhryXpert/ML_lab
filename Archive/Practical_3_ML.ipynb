## 1) Simple Linear Regression
Dataset: `placement.csv`  
Target: Predict package based on CGPA.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv("placement.csv")
df.head()
```

```python
df.info()
```

```python
# Scatter plot
plt.scatter(df["cgpa"], df["package"])
plt.xlabel("cgpa")
plt.ylabel("package (in lpa)")
plt.title("CGPA vs Package")
plt.show()
```

```python
# Train-test split
X = df[["cgpa"]]
y = df["package"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Train model
lr = LinearRegression()
lr.fit(X_train, y_train)

# Predict
y_pred = lr.predict(X_test)

print("R2 Score:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```

```python
# Regression line
plt.scatter(df["cgpa"], df["package"])
plt.plot(X_train, lr.predict(X_train), color="red")
plt.title("CGPA vs Package")
plt.xlabel("cgpa")
plt.ylabel("package (in lpa)")
plt.show()
```

## 2) Multi-Linear Regression
Dataset: `50_Startups.csv`  
Target: Predict `Profit` using `R&D Spend`, `Administration`, `Marketing Spend`, and `State`.

```python
df = pd.read_csv("50_Startups.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df = pd.get_dummies(df, columns=["State"], drop_first=True)
df.head()
```

```python
# Prepare features and target
X = df.drop("Profit", axis=1)
y = df["Profit"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Scaling is optional here, but can be used
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train model
mr = LinearRegression()
mr.fit(X_train_scaled, y_train)

# Predict
y_pred = mr.predict(X_test_scaled)

print("R2 Score:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```

## 3) Logistic Regression
Dataset: Iris dataset  
Target: Classify species from flower measurements.

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

iris = load_iris(as_frame=True)
df = iris.frame
df.head()
```

```python
# Pair plot
sns.pairplot(df, hue="target")
plt.show()
```

```python
# Features and target
X = df.drop("target", axis=1)
y = df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Train logistic regression
classifier = LogisticRegression(max_iter=1000)
classifier.fit(X_train, y_train)

# Predict
y_pred = classifier.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:", classification_report(y_test, y_pred))
print("Confusion Matrix:", confusion_matrix(y_test, y_pred))
```
