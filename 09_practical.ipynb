```python
import pandas as pd
import numpy as np
```

```python
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.estimators import MaximumLikelihoodEstimator
from pgmpy.inference import VariableElimination
from sklearn.preprocessing import KBinsDiscretizer
```

```python
data = pd.read_csv("src/heart.csv")
```

```python
continuous_cols = ['age','trestbps','chol','thalach']
```

```python
discratizer = KBinsDiscretizer(n_bins=3,encode='ordinal',strategy='uniform')
data[continuous_cols] = discratizer.fit_transform(data[continuous_cols])
```

```python
model = DiscreteBayesianNetwork([
    ('age','target'),
    ('sex','target'),
    ('cp','target'),
    ('trestbps','target'),
    ('chol','target'),
    ('fbs','target'),
    ('target','thalach')
])
```

```python
print(model)
```

```python
model.fit(data,estimator=MaximumLikelihoodEstimator)
```

```python
inference = VariableElimination(model)
```

```python
query = inference.query(
    variables = ['target'],
    evidence = {
        'age' : 2,
        'sex' : 1,
        'cp' : 3,
        'trestbps' : 1,
        'chol' : 2,
        'fbs' : 0,
        'thalach' : 1
    }
)
```

```python
print(query)
```
