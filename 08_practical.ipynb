```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
```

```python
data = pd.read_csv("src//different_clustering_data.csv")
print("Dataset:\n", data)
```

```python
X = data[['Feature1', 'Feature2']]
```

```python
gmm = GaussianMixture(n_components=3, random_state=42)
```

```python
gmm_labels = gmm.fit_predict(X)
```

```python
data['EM_Cluster'] = gmm_labels
```

```python
print("\nFinal Data with Cluster Labels:\n")
print(data)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42)
```

```python
kmeans_labels = kmeans.fit_predict(X)
```

```python
data['KMeans_Cluster'] = kmeans_labels
```

```python
print("\nFinal Data with Cluster Labels:\n")
print(data)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(X['Feature1'], X['Feature2'], c=kmeans_labels)
plt.title("K-Means Clustering")
plt.xlabel("Feature1")
plt.ylabel("Feature2")
plt.subplot(1,2,2)
plt.scatter(X['Feature1'], X['Feature2'], c=gmm_labels)
plt.title("EM (GMM) Clustering")
plt.xlabel("Feature1")
plt.ylabel("Feature2")
plt.show()
```
