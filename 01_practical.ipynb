```python
import statistics as st
```

```python
a= [10,20,50,10,21,90]
```

```python
mn = st.mean(a)
print("mean:",mn)
```

```python
md = st.mode(a)
print("Mode:",md)
```

```python
mdn = st.median(a)
print("Median",mdn)
```

```python
b = np.array([[10,20,30,50,10],[1,2,3,5,6]])
print(b)
print(b.ndim)
```

```python
b = np.array([[10,20,30,50,10],[1,2,3,5,6],[1,1,1,1,1]])
print(b)
```

```python
print("Dim:",b.ndim)
print("Shape:",b.shape)
```

```python
print("Size:",b.size)
print("Item Size:",b.itemsize)
print("Data type:",b.dtype)
```
